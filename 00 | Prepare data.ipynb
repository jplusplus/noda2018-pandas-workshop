```python
# Remove white spaces from column names
df_raw.columns = [x.strip() for x in df_raw.columns]

# Get a list of time columns
time_columns = df_raw.columns[1:].sort_values()
time_columns
```

```python
# Force conversion to numeric of value columns (this will take some time)
df_raw.loc[:, time_columns] = df_raw[time_columns].apply(pd.to_numeric, errors="coerce")
```

```python
# Split first column
index_columns = df_raw.columns[0].split(",") # ['unit', 'citizen', 'sex', 'age', 'decision', 'geo\\time']
index_columns[-1] = 'geo'

index = df_raw.iloc[:,0].str.split(",", expand=True)

index.columns = index_columns
index.head()
```

```python
# Remove original, unsplit column
df = df_raw.drop(df_raw.columns[0], axis=1)

# Merge 
df = pd.concat([index, df], axis=1)
df.head()
```

```python
# Inspect the index columns
df[index_columns].describe()
```

```python
# Get the 20 most common origins of applicants
top_countries = df.groupby("citizen")["2017Q4"].sum().sort_values(ascending=False).index[:20].tolist()
top_countries
```

```python
# Filter down the data
is_top_20 = df["citizen"].isin(top_countries)
is_total_age = df["age"] == "TOTAL"
is_total_sex = df["sex"] == "T"
is_selected_decisions = df["decision"].isin(["TOTAL", "TOTAL_POS"])

export_df = df[is_top_20 & is_total_age & is_total_sex & is_selected_decisions]
export_df.head()
```

```python
export_df = export_df.melt(id_vars=index_columns, value_vars=time_columns, var_name="time")\
               .drop(["unit", "sex","age"], axis=1)\
               .set_index(["citizen", "geo", "time", "decision"])\
               .unstack("decision").value\
               .reset_index()
                
export_df.head()
```

```python
file_path = "data/first instance decisions on applications.csv"
export_df.rename(columns={"geo": "country"}).set_index("citizen").to_csv(file_path, encoding="utf-8")
print("Save workshop data to {}".format(file_path))
```
